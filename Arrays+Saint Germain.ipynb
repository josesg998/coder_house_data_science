**PRÁCTICA CON ARRAYS EN NUMPY**

```python
#importamos la librería
import numpy as np
```



1. Generar un array aleatorio de 100 elementos. Calcular la mediana correspondiente.

```python
#Generamos un array aleatorio de 100 elementos
mi_array = np.random.randint(90,size=100)

print("Mi array es:\n\n",mi_array)

#Verificamos la cantidad de elementos del array
print(f"\nEste array tiene {np.size(mi_array)} elementos\n")

#Obtenemos la mediana del array
print("La mediana de mi array es: ",np.median(mi_array))
```



2. Generar un array que contenga siempre valor 1 en sus elementos y que tenga 20 elementos, luego otro array de rango 1 hasta el 20 inclusive. Sumarlos y luego decirme cual es el promedio del array resultante.



```python
#Generamos los dos arrays

## Array con 20 elementos igual a 1
mi_array_2 = np.ones(20,dtype=int)
## Array con un rango de 1 a 20
mi_array_3 = np.arange(1,21)

print("Array 1:\t",mi_array_2,"\n","\nArray 2:\t",mi_array_3)

#Sumamos ambos arrays
mi_array_4 = mi_array_2 + mi_array_3

print("\nSuma de Arrays:\t",mi_array_4)

#Calculamos su media
print("\nEl promedio de la suma de arrays es:\t",mi_array_4.mean())
```

3. En base al array del ajedrez generado en esta diapositivas , hacer un cambio de forma(reshape) y que sea de 4 filas y 16 columnas. Calcular su desvio estandar.

```python
#Generamos el array del ajedrez

ajedrez_64 = np.arange(1,65).reshape(8,8)

print("Array del ajedrez:\n\n",ajedrez_64)

#Realizamos un reshape para que tenga 4 filas y 16 columnas

ajedrez_reshape = ajedrez_64.reshape(4,16)

print("\nArray del ajedrez con reshape:\n\n",ajedrez_reshape)

# Calculamos el desvío estándar del array

print("\nEl desvío estándar del array del ajedrez con reshape es: ",np.std(ajedrez_reshape))
```
