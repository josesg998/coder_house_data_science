```python
#Importamos las librerías que necesitamos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Cargar este [archivo](https://drive.google.com/file/d/1Msf9fKLNyQJFmZbPecSe1G23kjkPg-ap/view) en Python. Realizar estadísticas descriptivas básicas.

```python
#Cargamos el archivo
df = pd.read_csv("/content/HRDataset_v14.csv")
#Miramos su forma
print(df.columns)
print("\nEl dataset tiene",df.shape[0],"filas y",df.shape[1],"columnas\n")
df.head()
```

```python
#Vemos estadísticas descriptivas
df.describe()
```

2. Realizar un histograma con los salarios ¿Qué rango(s) de salarios son los más populares?

```python
#Generamos el histograma
fig, ax = plt.subplots(figsize=(10,8))
ax.hist(df["Salary"])
ax.set_title("Distribución del salario",size=25)
ax.set_xlabel("Salario",size=15)
ax.set_ylabel("Frecuencia",size=15);
```

**El rango de salario más popular está entre 55,000 y 72,000**

3. Realizar un gráfico de violín con los salarios, discriminados por género.

```python
# Realizo el gráfico (como no encontré a qué género corresponde cada número, dejé la etiqueta del eje x sin modificaciones)
ax = sns.catplot(data=df, kind='violin', y='Salary',x="GenderID")
ax.set(xlabel='Género', ylabel='Salario', title='Distribución del salario por género');
```

4. Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire).

```python
#Transformo la variable en una serie de tiempo
df["DateofHire"] = pd.to_datetime(df["DateofHire"])
#Como es una sola variable, la forma que encontré de graficar esta variable es a partir de un histograma
fig, ax = plt.subplots(figsize=(16,8))
ax.hist(df["DateofHire"],edgecolor='black',bins=24)
ax.set_title("Contrataciones por año",size=25)
ax.set_ylabel("Contrataciones",size=20);
```
