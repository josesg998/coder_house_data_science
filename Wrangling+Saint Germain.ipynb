```python
#Cargamos Pandas
import pandas as pd
```

Consigna: Leer los set de datos w_data.csv y p_data.csv, fusionarlos a partir de la variable que tengan en común (ID) y realizar las siguientes ejercicios:


```python
#Cargamos las bases
w = pd.read_csv("w_data.csv",sep=";")
p = pd.read_csv("p_data.csv",sep=";")
```

```python
#Vemos la forma de los dataframe
print(f"El dataframe p tiene {p.shape[0]} filas y {p.shape[1]} columnas, mientras que el dataframe w tiene {w.shape[0]} filas y {w.shape[1]} columnas")
```

```python
#Unimos los bases
df = pd.merge(w,p,how="outer",on="ID")
print(f"El nuevo dataset tiene {df.shape[0]} filas y {df.shape[1]} columnas")
df.sample(5)
```

##2. Corregir los target a solo dos valores distintos ‘<=50k’ y ‘>50k’


```python
#Vemos que valores tiene la variable Target
df.Target.unique()
```

```python
# Modificamos los valores
df.loc[df.Target==" >50K.","Target"] = ">50K"
df.loc[df.Target==" >50K","Target"] = ">50K"
df.loc[df.Target==" <=50K.","Target"] = "<=50K"
df.loc[df.Target==" <=50K","Target"] = "<=50K"
```

```python
#Verificamos que se hayan modificado los valores
df.Target.unique()
```

##3. Calcular la proporción que hay entre personas con salarios >50k respecto del total para personas de distinto sexo, Raza y Educacion. Por ejemplo cuál es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y Educación


```python
# Armamos una tabla de frecuencia relativa e iteramos el comando para cada variable
for i in ["Sex","Race","Education"]:
  print(df[[i,"Target"]].value_counts(normalize=True).reset_index().pivot_table(index=i,columns="Target",values=0),"\n\n")
```
